**Remove garbage data**

```python
df = pd.read_csv("./data/domains_with_pricing_page.csv")

df = df[df["subscription_status"] == "subscription"].copy()
print("after subscription filter:", len(df))

df["pricing_url"] = (
    df["pricing_url"]
    .astype(str)
    .str.strip()
    .str.rstrip("/")
)

df = df[df["pricing_url"].notna() & (df["pricing_url"] != "nan")].copy()

missing_idxs = []
for idx, new_url in concise:
    if idx in df.index:
        df.at[idx, "pricing_url"] = str(new_url).strip().rstrip("/")
    else:
        missing_idxs.append(idx)

if missing_idxs:
    print("warning: indices not found:", missing_idxs)

remove_clean = [str(u).strip().rstrip("/") for u in remove]
mask_remove = df["pricing_url"].isin(remove_clean)
print("rows to remove:", int(mask_remove.sum()))

df = df.loc[~mask_remove].copy()
print("after remove:", len(df))

#remove duplicates (keep first)
df = df.drop_duplicates(subset=["pricing_url"], keep="first")
df = df.drop_duplicates(subset=["domain"], keep="first")

# columns I believe are important to keep
keep_cols = [
    "domain_normalized",
    "domain",
    "pricing_url",
    "pricing_url_method",
    "dynamic_components",
    "popup_overlay",
    "notes",
]
df = df[[c for c in keep_cols if c in df.columns]].copy()

if "notes" not in df.columns:
    df["notes"] = ""

def append_note(series: pd.Series, note: str) -> pd.Series:
    s = series.fillna("").astype(str).str.strip()
    return s.where(s == "", s + " | ") + note

# mark physical subscriptions to Notes
physical_set = {str(u).strip().rstrip("/") for u in physical}
mask_physical = df["pricing_url"].isin(physical_set)
df.loc[mask_physical, "notes"] = append_note(
    df.loc[mask_physical, "notes"],
    "physical magazine/newspaper subscription"
)

# mark API pricing pages to Notes
api_set = {str(u).strip().rstrip("/") for u in api}
mask_api = df["pricing_url"].isin(api_set)
df.loc[mask_api, "notes"] = append_note(
    df.loc[mask_api, "notes"],
    "API pricing page"
)
```
